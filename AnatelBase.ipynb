```python
tipo_prepos = pd.read_csv('Movel_Pessoal/Por_Tipo/csv/Acessos_SMP_2015-2018_-_Tipo.csv', sep=';', encoding= 'latin-1' )
```

```python
tipo_prepos.head(1)
```

```python
#criando um dataframe para guardar as medias de pré-pago em forma de linha/coluna.
pre_pago= pd.DataFrame()

#criando um dataframe para guardar as medias de pós-pago em forma de linha/coluna
pos_pago= pd.DataFrame()

#dividindo em dois dataframes, pré e pós.
pre = tipo_prepos.query('Tipo == "Pré-pago"')
pos = tipo_prepos.query('Tipo == "Pós-pago"')



#Somando os valores entre Janeiro e Maio dos anos de 2015 ate 2018, separados pelos seus respectivos anos

p_soma2015= pre['2015-01']+pre['2015-02']+pre['2015-03']+pre['2015-04']+pre['2015-05']
p_soma2016= pre['2016-01']+pre['2016-02']+pre['2016-03']+pre['2016-04']+pre['2016-05']
p_soma2017= pre['2017-01']+pre['2017-02']+pre['2017-03']+pre['2017-04']+pre['2017-05']
p_soma2018= pre['2018-01']+pre['2018-02']+pre['2018-03']+pre['2018-04']+pre['2018-05']


#Para facilitar na compreensão gráfica dos valores, foram feitas as seguintes divisões, a fim de simplificar.
p_soma2015 = p_soma2015/1000 
p_soma2016= p_soma2016/1000
p_soma2017= p_soma2017/1000
p_soma2018= p_soma2018/1000

pre_pago['total2015_ate05']=p_soma2015
pre_pago['total2016_ate05']=p_soma2016
pre_pago['total2017_ate05']=p_soma2017
pre_pago['total2018_ate05']=p_soma2018


pre_medias = pd.concat([pre,pre_pago],axis=1)
pre_medias

```

```python
pre_medias.groupby('Tipo')['total2015_ate05','total2016_ate05','total2017_ate05','total2018_ate05'].head(20).plot(kind='bar',figsize=(13,5), grid=True, rot=0) 

plt.title("Total de Acessos no Pré-Pago no Brasil período de Janeiro a Maio")
plt.xlabel("")
plt.ylabel("Quantidade de Acessos")
plt.show()
```

```python
pos_soma2015= pos['2015-01']+pos['2015-02']+pos['2015-03']+pos['2015-04']+pos['2015-05']
pos_soma2016= pos['2016-01']+pos['2016-02']+pos['2016-03']+pos['2016-04']+pos['2016-05']
pos_soma2017= pos['2017-01']+pos['2017-02']+pos['2017-03']+pos['2017-04']+pos['2017-05']
pos_soma2018= pos['2018-01']+pos['2018-02']+pos['2018-03']+pos['2018-04']+pos['2018-05']


#Para facilitar na compreensão gráfica dos valores, foram feitas as seguintes divisões, a fim de simplificar.
pos_soma2015 = pos_soma2015/100
pos_soma2016= pos_soma2016/100
pos_soma2017= pos_soma2017/100
pos_soma2018= pos_soma2018/100

pos_pago['total2015_ate05']=pos_soma2015
pos_pago['total2016_ate05']=pos_soma2016
pos_pago['total2017_ate05']=pos_soma2017
pos_pago['total2018_ate05']=pos_soma2018


pos_medias = pd.concat([pos,pos_pago],axis=1)
pos_medias
```
